```python
import pandas as pd
import sqlite3

mat = pd.read_csv('student-mat.csv')
por = pd.read_csv('student-por.csv')
```

```python
# Add a student_id column to 'mat'
mat.insert(0, 'student_id', range(1000, 1000 + len(mat)))
print(mat.tail())
```

```python
# Add a student_id column to 'por'
por.insert(0, 'student_id', range(1000 + len(mat), 1000 + len(mat) + len(por)))
print(por.head())
```

```python
# Concatenate 'mat' and 'por'
combined_df = pd.concat([mat,por], ignore_index=True)
```

```python
combined_df
```

```python
# only keep fields needed for analysis objectives
analysis_df = combined_df[['student_id','sex','age','Pstatus','studytime','failures','activities','paid','romantic','higher','famrel','freetime','goout','Dalc','Walc','health','absences','G1','G2','G3']]
analysis_df.rename(columns={'higher':'higher_edu','famrel':'family_relationship','freetime':'free_time','paid':'tutored','goout':'go_out','Dalc':'workday_alcohol','Walc':'weekend_alcohol','romantic':'currently_dating','health':'health_status','G1':'s1_grade','G2':'s2_grade','G3':'final_grade','Pstatus':'parent_status','studytime':'study_time'}, inplace=True)
```

```python
# Portugal has a different grading scale, so convert grades from 0-20 to A-F grading scale
def grade_conversion(grade):
    if grade >= 18:
        return 'A+'
    elif grade >= 16:
        return 'A'
    elif grade >= 14:
        return 'B'
    elif grade >= 10:
        return 'C'
    else:
        return 'F'
    
analysis_df['s1_grade'] = analysis_df['s1_grade'].apply(grade_conversion)
analysis_df['s2_grade'] = analysis_df['s2_grade'].apply(grade_conversion)
analysis_df['final_grade'] = analysis_df['final_grade'].apply(grade_conversion)
```

```python
analysis_df
```

```python
final_grade_counts = analysis_df['final_grade'].value_counts()
print(final_grade_counts)
```

```python
# Split columns into structured tables for SQL
student_df = analysis_df[['student_id','sex','age','parent_status','study_time','failures','activities','higher_edu','family_relationship','currently_dating','tutored','free_time','go_out','workday_alcohol','weekend_alcohol','health_status','absences']]
grades_df = analysis_df[['student_id','s1_grade','s2_grade','final_grade']]
```

```python
# create and connect to a sqlite database
connection = sqlite3.connect('student_alcohol_data.db')

# create a cursor oboject to execute SQL commands
cursor = connection.cursor()

# SQL statement to create a table with a primary key
create_student_table_query = """
# CREATE TABLE IF NOT EXISTS "student" (
# 	"student_id" INTEGER NOT NULL UNIQUE,
# 	"sex" VARCHAR NOT NULL,
# 	"age" INTEGER NOT NULL,
# 	"parent_status" VARCHAR,
# 	"study_time" INTEGER,
# 	"failures" INTEGER,
# 	"activities" VARCHAR,
# 	"higher_edu" VARCHAR,
# 	"family_relationship" INTEGER,
# 	"free_time" INTEGER,
# 	"go_out" INTEGER,
# 	"workday_alcohol" INTEGER,
# 	"weekend_alcohol" INTEGER,
# 	"health_status" INTEGER,
# 	"absences" INTEGER,
# 	"currently_dating" VARCHAR,
# 	"tutored" VARCHAR,
# 	PRIMARY KEY("student_id"),
# 	FOREIGN KEY ("student_id") REFERENCES "grades"("student_id")
# 	ON UPDATE NO ACTION ON DELETE NO ACTION
# );
# """

create_grades_table_query = """
CREATE TABLE IF NOT EXISTS "grades" (
	"student_id" INTEGER NOT NULL UNIQUE,
	"s1_grade" VARCHAR,
	"s2_grade" VARCHAR,
	"final_grade" VARCHAR NOT NULL,
	PRIMARY KEY("student_id")
);
"""

# Execute create table queries
cursor.execute(create_student_table_query)
cursor.execute(create_grades_table_query)

# write dataframe records to a table in SQLite database
try:
	student_df.to_sql(
		name='student',       # name of the table in the database
        con=connection,     # database connection to use
        if_exists='append', # append data to the existing table (vs. replace)
        index=False         # do not write the DataFrame index as a column
    )

	grades_df.to_sql(
		name='grades',       # name of the table in the database
        con=connection,     # database connection to use
        if_exists='append', # append data to the existing table (vs. replace)
        index=False         # do not write the DataFrame index as a column
    )

	print("Data inserted successfully!")
except sqlite3.IntegrityError as e:
    print(f"Integrity error occurred: {e}")
finally:
	# close the connection
	connection.close()
```
